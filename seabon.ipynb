```python
1:Plot | Function | Use
1. Line Plot | sns.lineplot() | Shows trends or changes over time for
continuous data.
2. Bar Plot | sns.barplot() | Compares average values across categories
with confidence intervals.
3. Box Plot | sns.boxplot() | Visualizes the distribution, median, and
outliers of data.
4. Scatter Plot | sns.scatterplot() | Displays relationships or
correlations between two numeric variables.
5. Histogram (Displot) | sns.histplot() or sns.displot() |
Shows the distribution of a single variable (like frequency of values).

2:import seaborn as sns
import matplotlib.pyplot as plt

# Load the fMRI dataset
fmri = sns.load_dataset('fmri')

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region", markers=True)

# Set plot labels and title
plt.title("fMRI Signal over Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event & Region")

# Display the plot
plt.show()

3:import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a figure for the box plots
plt.figure(figsize=(12, 6))

# First box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title('Box plot of Age by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Second box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title('Box plot of Fare by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Fare')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()


4:import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Plot a histogram for the 'price' column with hue based on 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', kde=True, multiple='stack')

# Set plot labels and title
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Display the plot
plt.show()


5:import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Plot a pair plot with hue based on 'species'
sns.pairplot(iris, hue='species')

# Display the plot
plt.show()

6:import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Pivot the data for the heatmap (month vs. year)
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, cmap="YlGnBu", fmt="d")

# Set plot title
plt.title('Heatmap of Flights Passengers by Month and Year')

# Display the plot
plt.show()
```
