```python
1:import pandas as pd

# Load the dataset
df = pd.read_csv("flight_price.csv")  # Replace with the actual filename

# Display the first few rows
print(df.head())

# Get the shape of the dataset
rows, columns = df.shape
print(f"The dataset has {rows} rows and {columns} columns.")

2:import matplotlib.pyplot as plt
import seaborn as sns

# Check if 'Price' column exists
print(df['Price'].describe())  # Summary statistics

# Plot histogram of flight prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


3:# Summary statistics of the Price column
min_price = df['Price'].min()
max_price = df['Price'].max()
price_range = max_price - min_price

print(f"Minimum Price: ₹{min_price}")
print(f"Maximum Price: ₹{max_price}")
print(f"Price Range: ₹{price_range}")

4:import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size and style
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")

# Create the boxplot
sns.boxplot(x='Airline', y='Price', data=df, palette='Set3')

# Improve plot readability
plt.xticks(rotation=45)
plt.title('Flight Price Distribution by Airline')
plt.xlabel('Airline')
plt.ylabel('Price (INR)')
plt.tight_layout()
plt.show()


5:import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price'], color='lightcoral')
plt.title('Boxplot of Flight Prices - Detecting Outliers')
plt.xlabel('Price (INR)')
plt.show()

6:df['Journey_Date'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
df['Month'] = df['Journey_Date'].dt.month
df['Day_of_Week'] = df['Journey_Date'].dt.day_name()
monthly_prices = df.groupby('Month')['Price'].mean()
monthly_counts = df['Month'].value_counts().sort_index()
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10,6))

ax1.plot(monthly_prices.index, monthly_prices.values, color='red', marker='o')
ax1.set_ylabel('Average Price', color='red')
ax1.set_xlabel('Month')

ax2 = ax1.twinx()
ax2.bar(monthly_counts.index, monthly_counts.values, alpha=0.3, color='blue')
ax2.set_ylabel('Number of Bookings', color='blue')

plt.title('Flight Price and Booking Volume by Month')
plt.show()

7:df['Journey_Date'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')
df['Month'] = df['Journey_Date'].dt.month

monthly_avg_price = df.groupby('Month')['Price'].mean()

plt.figure(figsize=(10,5))
plt.plot(monthly_avg_price.index, monthly_avg_price.values, marker='o', linestyle='-')
plt.title('Average Flight Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Price (INR)')
plt.grid(True)
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(x='Airline', y='Price', data=df)
plt.xticks(rotation=45)
plt.title('Flight Price Distribution by Airline')
plt.show()
df['Booking_Date'] = pd.to_datetime(df['Booking_Date'])  # if available
df['Lead_Time'] = (df['Journey_Date'] - df['Booking_Date']).dt.days

plt.figure(figsize=(10,6))
sns.scatterplot(x='Lead_Time', y='Price', data=df, alpha=0.3)
plt.title('Price vs Lead Time')
plt.xlabel('Days before Journey')
plt.ylabel('Price (INR)')
plt.show()

8:import seaborn as sns

numeric_cols = ['Price', 'Duration', 'Lead_Time']  # add numeric features
corr = df[numeric_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between Numeric Features')
plt.show()
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
le = LabelEncoder()
df['Airline_enc'] = le.fit_transform(df['Airline'])

features = ['Airline_enc', 'Duration', 'Total_Stops', 'Lead_Time']  # example features
X = df[features]
y = df['Price']

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

importances = model.feature_importances_
feature_importance = pd.Series(importances, index=features).sort_values(ascending=False)

feature_importance.plot(kind='bar', color='salmon')
plt.title('Feature Importance for Flight Price Prediction')
plt.show()


9:import pandas as pd

# Load the dataset
df = pd.read_csv('googleplaystore.csv')  # Replace with your file path

# Display first few rows to confirm loading
print(df.head())

# Get dimensions of the dataset
rows, columns = df.shape
print(f"The dataset has {rows} rows and {columns} columns.")

10:import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your loaded Google Playstore dataframe

plt.figure(figsize=(14, 7))
sns.boxplot(x='Category', y='Rating', data=df, palette='Set3')

plt.xticks(rotation=45, ha='right')  # Rotate category names for readability
plt.title('Distribution of App Ratings by Category')
plt.xlabel('App Category')
plt.ylabel('App Rating')
plt.ylim(0, 5)  # Ratings typically range from 0 to 5

plt.tight_layout()
plt.show()

11:# Check for missing values in each column
missing_values = df.isnull().sum()

# Filter columns with missing values only
missing_values = missing_values[missing_values > 0]

print("Columns with missing values and their counts:")
print(missing_values)


12:import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Clean 'Size' column: convert sizes like '19M', '14k' to numeric values in MB

def size_to_mb(size):
    if isinstance(size, str):
        if size.endswith('M'):
            return float(size[:-1])
        elif size.endswith('k'):
            return float(size[:-1]) / 1024
        else:
            return np.nan  # For 'Varies with device' or other strings
    return np.nan

df['Size_MB'] = df['Size'].apply(size_to_mb)

# Drop rows with missing Size_MB or Rating
clean_df = df.dropna(subset=['Size_MB', 'Rating'])

plt.figure(figsize=(10,6))
sns.scatterplot(x='Size_MB', y='Rating', data=clean_df, alpha=0.5)

plt.title('Scatter Plot of App Size vs. Rating')
plt.xlabel('App Size (MB)')
plt.ylabel('App Rating')
plt.xlim(0, clean_df['Size_MB'].quantile(0.99))  # Limit x-axis to 99th percentile to avoid extreme outliers
plt.ylim(0, 5)

plt.show()


13:import matplotlib.pyplot as plt
import seaborn as sns

# Clean the Price column: remove currency symbols and convert to float
df['Price_Clean'] = df['Price'].str.replace('$', '').astype(float)

# Calculate average price by app Type
avg_price_by_type = df.groupby('Type')['Price_Clean'].mean().reset_index()

plt.figure(figsize=(7,5))
sns.barplot(x='Type', y='Price_Clean', data=avg_price_by_type, palette='pastel')

plt.title('Average App Price by Type')
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.ylim(0, avg_price_by_type['Price_Clean'].max() + 5)

plt.show()

14:import pandas as pd

# Clean 'Installs' column: remove commas and plus signs, convert to int
df['Installs_Clean'] = df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Group by app name and sum installs (if multiple entries per app)
installs_by_app = df.groupby('App')['Installs_Clean'].sum()

# Sort descending and get top 10
top_10_apps = installs_by_app.sort_values(ascending=False).head(10)

# Display frequency table
print("Top 10 Most Popular Apps by Number of Installs:")
print(top_10_apps)

15:# Clean installs column
df['Installs_Clean'] = df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Group data by category
category_stats = df.groupby('Category').agg({
    'Installs_Clean': 'sum',
    'App': 'count',
    'Rating': 'mean',
    'Reviews': 'mean'
}).rename(columns={
    'Installs_Clean': 'Total_Installs',
    'App': 'Number_of_Apps',
    'Rating': 'Average_Rating',
    'Reviews': 'Average_Reviews'
}).reset_index()

# Sort by total installs descending
category_stats_sorted = category_stats.sort_values(by='Total_Installs', ascending=False)

print(category_stats_sorted.head(10))

16:# Group by Developer (assuming 'Developer' column exists)
developer_stats = df.groupby('Developer').agg({
    'App': 'count',
    'Installs_Clean': 'sum',
    'Rating': 'mean',
    'Reviews': 'sum'
}).rename(columns={
    'App': 'Number_of_Apps',
    'Installs_Clean': 'Total_Installs',
    'Rating': 'Average_Rating',
    'Reviews': 'Total_Reviews'
}).reset_index()

# Filter developers with a minimum number of apps (e.g., 3) to focus on active developers
active_developers = developer_stats[developer_stats['Number_of_Apps'] >= 3]

# Sort by total installs descending
top_developers = active_developers.sort_values(by='Total_Installs', ascending=False).head(10)

print(top_developers)


17:# Convert 'Last Updated' to datetime
df['Last_Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')

# Extract month and year
df['Launch_Month'] = df['Last_Updated'].dt.month
df['Launch_Year'] = df['Last_Updated'].dt.year

# Group by month to analyze installs
monthly_installs = df.groupby('Launch_Month')['Installs_Clean'].mean()

# Plot average installs by launch month
monthly_installs.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Average Installs by Launch Month')
plt.xlabel('Month')
plt.ylabel('Average Number of Installs')
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```
