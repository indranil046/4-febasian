```python
1:import scipy.stats as stats
import math

# Given values
mean = 50
std_dev = 5
n = 30  # Assuming sample size
confidence_level = 0.95

# Standard Error
SE = std_dev / math.sqrt(n)

# Z-value for 95% confidence level
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Margin of Error
margin_of_error = z_value * SE

# Confidence Interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

lower_bound, upper_bound

2:Since the p-value (0.9626) is much greater than the significance level
we fail to reject the null hypothesis. This means that there is no significant
difference between the observed distribution and the expected distribution of
M&M colors. In other words, the distribution of M&M colors in the bag appears
to match the expected distribution.

3:import numpy as np

# Contingency table data
data = np.array([[20, 15],
                 [10, 25],
                 [15, 20]])

# Perform the chi-square test for independence
chi2_stat_contingency, p_value_contingency, dof, expected = stats.chi2_contingency(data)

chi2_stat_contingency, p_value_contingency, dof, expected

The p-value (0.0541) is slightly greater than the significance level of 0.05.

This means that we fail to reject the null hypothesis at the 5% significance
level, which suggests that there is no significant association between the groups
and the outcomes. The observed frequencies are not significantly different from
the expected frequencies under the assumption of independence.

4:# Given values
n = 500  # sample size
smokers = 60  # number of individuals who smoke
confidence_level = 0.95

# Sample proportion
p_hat = smokers / n

# Z-value for 95% confidence level
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error for proportion
SE = math.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of error
margin_of_error = z_value * SE

# Confidence Interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

lower_bound, upper_bound

5:# Given values
mean = 75
std_dev = 12
n = 30  # Assuming sample size
confidence_level = 0.90

# Standard Error
SE = std_dev / math.sqrt(n)

# Z-value for 90% confidence level
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Margin of Error
margin_of_error = z_value * SE

# Confidence Interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

lower_bound, upper_bound

6:import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Generate chi-square distribution data
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})', color='blue')

# Shade the area corresponding to chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
y_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label='Area under curve (Chi-Square <= 15)')

# Add labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom', fontsize=14)
plt.xlabel('Chi-Square Value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.axvline(x=15, color='red', linestyle='--', label='Chi-Square Statistic = 15')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

7:# Given values
n = 1000  # sample size
coke_preference = 520  # number of people who prefer Coke
confidence_level = 0.99

# Sample proportion
p_hat = coke_preference / n

# Z-value for 99% confidence level
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error for proportion
SE = math.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of error
margin_of_error = z_value * SE

# Confidence Interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

lower_bound, upper_bound

8:# Observed frequencies
observed_tails = 45
observed_heads = 55

# Expected frequencies
expected_tails = 50
expected_heads = 50

# Chi-square calculation
chi_square_stat = ((observed_tails - expected_tails)**2 / expected_tails) + ((observed_heads - expected_heads)**2 / expected_heads)
chi_square_stat

9:# Given data
observed = {
    "Smoker_Yes": 60,
    "Smoker_No": 140,
    "Non_smoker_Yes": 30,
    "Non_smoker_No": 170
}

# Row totals
total_smoker = observed["Smoker_Yes"] + observed["Smoker_No"]
total_non_smoker = observed["Non_smoker_Yes"] + observed["Non_smoker_No"]
# Column totals
total_yes = observed["Smoker_Yes"] + observed["Non_smoker_Yes"]
total_no = observed["Smoker_No"] + observed["Non_smoker_No"]
# Grand total
grand_total = total_smoker + total_non_smoker

# Expected frequencies
expected_Smoker_Yes = (total_smoker * total_yes) / grand_total
expected_Smoker_No = (total_smoker * total_no) / grand_total
expected_Non_smoker_Yes = (total_non_smoker * total_yes) / grand_total
expected_Non_smoker_No = (total_non_smoker * total_no) / grand_total

# Chi-square statistic calculation
chi_square_stat = (
    ((observed["Smoker_Yes"] - expected_Smoker_Yes) ** 2) / expected_Smoker_Yes +
    ((observed["Smoker_No"] - expected_Smoker_No) ** 2) / expected_Smoker_No +
    ((observed["Non_smoker_Yes"] - expected_Non_smoker_Yes) ** 2) / expected_Non_smoker_Yes +
    ((observed["Non_smoker_No"] - expected_Non_smoker_No) ** 2) / expected_Non_smoker_No
)

# Results
chi_square_stat, expected_Smoker_Yes, expected_Smoker_No, expected_Non_smoker_Yes, expected_Non_smoker_No

```
