```python
1:import numpy as np
from scipy.stats import pearsonr

# Sample data: time spent studying (hours) and exam scores (out of 100)
study_time = [2, 4, 3, 5, 6, 8, 7, 3, 5, 6]
exam_scores = [50, 65, 60, 70, 75, 85, 80, 60, 70, 75]

# Calculate Pearson correlation coefficient and p-value
corr_coef, p_value = pearsonr(study_time, exam_scores)

print(f"Pearson correlation coefficient: {corr_coef:.2f}")
print(f"P-value: {p_value:.4f}")

2:from scipy.stats import spearmanr

# Sample data: hours of sleep and job satisfaction (scale 1-10)
sleep_hours = [6, 7, 5, 8, 7, 6, 9, 5, 7, 6]
job_satisfaction = [7, 8, 6, 9, 8, 7, 10, 5, 8, 7]

# Calculate Spearman's rank correlation and p-value
spearman_corr, p_value = spearmanr(sleep_hours, job_satisfaction)

print(f"Spearman's rank correlation coefficient: {spearman_corr:.2f}")
print(f"P-value: {p_value:.4f}")

3:import numpy as np
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt

# Simulate data (for demonstration)
np.random.seed(0)
exercise_hours = np.random.uniform(0, 10, 50)  # Hours of exercise per week
# Assume BMI negatively correlates with exercise but with some noise
bmi = 30 - 0.8 * exercise_hours + np.random.normal(0, 2, 50)

# Calculate Pearson correlation coefficient
pearson_corr, pearson_p = pearsonr(exercise_hours, bmi)

# Calculate Spearman's rank correlation coefficient
spearman_corr, spearman_p = spearmanr(exercise_hours, bmi)

print(f"Pearson correlation coefficient: {pearson_corr:.2f} (p-value={pearson_p:.4f})")
print(f"Spearman's rank correlation coefficient: {spearman_corr:.2f} (p-value={spearman_p:.4f})")

# Optional: scatter plot with regression line
plt.scatter(exercise_hours, bmi, label='Data points')
plt.xlabel('Hours of Exercise per Week')
plt.ylabel('BMI')
plt.title('Exercise Hours vs BMI')
plt.show()

4:import numpy as np
from scipy.stats import pearsonr

# Example data for 50 participants
tv_hours = np.array([
    2, 3, 1.5, 4, 5, 2.5, 3, 6, 1, 4, 2, 3.5, 5, 4.5, 2, 3, 1, 5.5, 4, 3,
    2, 4, 3, 1, 6, 2.5, 3.5, 4, 5, 1, 3, 4, 2, 5.5, 6, 1, 2, 3, 5, 4, 3,
    2, 4, 3, 1, 5, 2, 3, 4, 1, 5
])

physical_activity = np.array([
    5, 4, 6, 3, 2, 5, 4, 2, 7, 3, 5, 4, 2, 3, 5, 4, 6, 1, 3, 4,
    5, 3, 4, 7, 2, 5, 4, 3, 2, 6, 4, 3, 5, 2, 1, 6, 5, 4, 2, 3,
    5, 3, 4, 7, 2, 5, 4, 3, 6, 2
])

# Calculate Pearson correlation coefficient
corr_coef, p_value = pearsonr(tv_hours, physical_activity)

print(f"Pearson correlation coefficient: {corr_coef:.2f}")
print(f"P-value: {p_value:.4f}")


5:import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Create dataframe
data = {
    'Age': [25, 42, 37, 19, 31, 28],
    'Brand': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']
}

df = pd.DataFrame(data)

# Boxplot to visualize age by brand
plt.figure(figsize=(8, 5))
sns.boxplot(x='Brand', y='Age', data=df)
plt.title('Age Distribution by Soft Drink Brand Preference')
plt.show()

# ANOVA test to check if mean age differs by brand
groups = [df[df['Brand'] == brand]['Age'] for brand in df['Brand'].unique()]
f_stat, p_value = f_oneway(*groups)

print(f"ANOVA F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")


6:import numpy as np
from scipy.stats import pearsonr

# Simulate example data for 30 sales representatives
np.random.seed(42)
sales_calls_per_day = np.random.randint(5, 20, size=30)  # Calls per day between
 5 and 20
# Assuming sales made per week increases with calls per day, with some noise
sales_made_per_week = sales_calls_per_day * 3 + np.random.normal(0, 5, 30)

# Calculate Pearson correlation coefficient
corr_coef, p_value = pearsonr(sales_calls_per_day, sales_made_per_week)

print(f"Pearson correlation coefficient: {corr_coef:.2f}")
print(f"P-value: {p_value:.4f}")
The very low p-value (< 0.05) suggests the correlation is statistically
significant, unlikely due to random chance.
```
