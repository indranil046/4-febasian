```python
1:The Probability Density Function (PDF) is a function that describes the
relative likelihood of a continuous random variable taking on a particular value.


2:There are two main types of probability distributions, based on the type of
random variable involved:

1. Discrete Probability Distributions
These deal with discrete random variables (variables that can take on countable
                                           values).

Common types:
Binomial Distribution: Probability of a fixed number of successes in a fixed
number of independent Bernoulli trials (e.g., coin tosses).

Poisson Distribution: Models the number of events occurring in a fixed interval
of time or space (e.g., number of emails per hour).

Geometric Distribution: Models the number of trials until the first success.

Bernoulli Distribution: A single trial with two possible outcomes (success/failure).

2. Continuous Probability Distributions
These deal with continuous random variables (variables that can take on any
                                             value within a range).

Common types:
Normal Distribution (Gaussian): Bell-shaped, symmetric about the mean.

Exponential Distribution: Models the time between independent events that happen
 at a constant average rate.

Uniform Distribution: All values in a given range are equally likely.

Beta & Gamma Distributions: Useful in Bayesian statistics and modeling waiting
times.


3:import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density of the normal distribution at point x.

    Parameters:
    - x: Value at which to evaluate the PDF
    - mean: Mean of the normal distribution
    - std_dev: Standard deviation of the distribution

    Returns:
    - PDF value at x
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    return coefficient * math.exp(exponent)

# Example usage:
print(normal_pdf(0, mean=0, std_dev=1))  # Standard normal PDF at x=0

4:The binomial distribution is a discrete probability distribution that
describes the number of successes in a fixed number of independent Bernoulli
trials (each with two possible outcomes: success or failure).

5:import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10            # number of trials
p = 0.4           # probability of success
size = 1000       # sample size

# Generate random sample
sample = np.random.binomial(n=n, p=p, size=size)

# Plot histogram
plt.hist(sample, bins=range(n+2), edgecolor='black', align='left')
plt.title('Histogram of Binomial Distribution (n=10, p=0.4)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

6:from scipy.stats import poisson

# CDF at x=3 with mean Î»=2.5
cdf_value = poisson.cdf(3, mu=2.5)
print(cdf_value)

7:Use Binomial when the number of trials is fixed and you're counting successes.

Use Poisson when events occur randomly over time/space and you're
 counting occurrences.

8:import numpy as np

# Parameters
lambda_val = 5
sample_size = 1000

# Generate random sample
sample = np.random.poisson(lam=lambda_val, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)


9:the mean and variance are equal in a Poisson distribution.
This is a defining property.

10:The least frequent data is found at the farthest points from the mean in a
normal distribution, specifically in the tails where the probability density is
smallest.
```
