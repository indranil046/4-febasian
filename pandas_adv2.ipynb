```python
1:import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]

df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})
print(df.iloc[1])


2:df.loc[] — Label-based indexing
 df.iloc[] — Position-based indexing

3:import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating the original DataFrame
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

reindex = [3, 0, 1, 2]
new_df = df1.reindex(reindex)

new_df.loc[2] tries to access the row with index label 2.

new_df.iloc[2] tries to access the third row (position index 2) in new_df.

4:# (i) Mean of each column
column_means = df1.mean()

# (ii) Standard deviation of column 'column_2'
std_column_2 = df1['column_2'].std()

# Displaying the results
print("Mean of each column:\n", column_means)
print("\nStandard Deviation of 'column_2':", std_column_2)



5:# Replace second row of 'column_2' with a string
df1.loc[2, 'column_2'] = 'hello'
mean_col2 = df1['column_2'].mean()
print("Mean of column_2:", mean_col2)
TypeError: could not convert string to float: 'hello'
mean_col2 = pd.to_numeric(df1['column_2'], errors='coerce').mean()
print("Mean of column_2 (ignoring non-numeric):", mean_col2)


6:df['column'].rolling(window=3).mean()


7:import pandas as pd

# Get current datetime
current_time = pd.Timestamp.now()

# Print current month and year
print("Current Month:", current_time.month)
print("Current Year:", current_time.year)


8:import pandas as pd

# Get input from the user
date1_str = input("Enter the first date (YYYY-MM-DD): ")
date2_str = input("Enter the second date (YYYY-MM-DD): ")

# Convert strings to pandas Timestamps
date1 = pd.to_datetime(date1_str)
date2 = pd.to_datetime(date2_str)

# Calculate time difference
time_diff = abs(date2 - date1)  # use abs to always get a positive difference

# Extract days, hours, and minutes
days = time_diff.days
hours = time_diff.seconds // 3600
minutes = (time_diff.seconds % 3600) // 60

# Display results
print(f"\nDifference between the dates:")
print(f"{days} days, {hours} hours, and {minutes} minutes")



9:import pandas as pd

# Read the CSV file
file_path = input("Enter the CSV file path: ")
df = pd.read_csv(file_path)

# Display columns for user reference
print("\nAvailable columns in the CSV file:")
print(df.columns)

# Get the column name and category order from the user
column_name = input("\nEnter the column name to convert to categorical: ")
category_order = input(f"Enter the category order for '{column_name}' (comma separated): ").split(',')

# Convert the specified column to categorical with the given order
df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

# Sort the data by the categorical column
sorted_df = df.sort_values(by=column_name)

# Display the sorted data
print("\nSorted data based on the specified categorical column:")
print(sorted_df)


10:import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = input("Enter the CSV file path: ")
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(df.head())

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Pivot the dataframe to get sales by product category and date
pivot_df = df.pivot_table(index='Date', columns='ProductCategory', values='Sales', aggfunc='sum', fill_value=0)

# Plotting the stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')

# Adding titles and labels
plt.title('Sales by Product Category Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)

# Display the chart
plt.tight_layout()
plt.show()


11:import pandas as pd

# Prompt the user to enter the file path of the CSV file
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# Calculate the mean, median, and mode of the 'Test Score' column
mean_score = df['Test Score'].mean()
median_score = df['Test Score'].median()
mode_score = df['Test Score'].mode().tolist()  # Get the mode as a list (in case there are multiple modes)

# Display the results in a table
result_table = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Value': [mean_score, median_score, ', '.join(map(str, mode_score))]
})

# Print the result table
print("\n+-----------+--------+")
print(result_table.to_string(index=False, header=True))
print("+-----------+--------+")
```
