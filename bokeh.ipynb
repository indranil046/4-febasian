```python
1:from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display the plot inline in a notebook (or use output_file("plot.html") for file)
output_notebook()

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Simple Line Plot", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Add a line renderer
p.line(x, y, legend_label="Line", line_width=2)

# Show the plot
show(p)


2:Glyphs in Bokeh are the visual building blocks of a plot — they are the shapes
 or marks used to represent your data, such as:

Line

Circle

Square

Triangle

Rect (bar)

Text

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 2, 7]

# Create a figure
p = figure(title="Line and Circle Glyphs", x_axis_label='X', y_axis_label='Y')

# Add a line glyph
p.line(x, y, line_width=2, color="blue", legend_label="Line")

# Add a circle glyph at data points
p.circle(x, y, size=10, color="red", legend_label="Points")

# Show plot
show(p)

3:title
from bokeh.plotting import figure, show

p = figure(title="My Customized Plot")
p.title.text_font_size = '16pt'
p.title.align = 'center'
p.title.text_color = 'navy'
p.title.text_font = 'times'

Customizing the Axes
p.xaxis.axis_label = "X-Axis Label"
p.yaxis.axis_label = "Y-Axis Label"

p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"

p.xaxis.major_label_text_color = "green"
p.yaxis.major_label_text_color = "red"

p.xaxis.major_tick_line_color = "black"
p.yaxis.minor_tick_line_color = "gray"

Customizing the Legend
p.line([1, 2, 3], [4, 6, 2], legend_label="Sample Line", line_color="blue")

p.legend.title = "Legend"
p.legend.label_text_font_size = '10pt'
p.legend.label_text_color = "black"
p.legend.background_fill_color = "#f5f5f5"
p.legend.border_line_color = "black"
p.legend.location = "top_left"

4:A Bokeh server is a powerful tool that allows you to create interactive web
applications with Python callbacks. Unlike standalone HTML/JS plots, a Bokeh
server app keeps a live connection between the browser and a Python process
running on the server. This means your plot can respond to user input or update
in real time using Python code.

5:a:Create a Bokeh plot in Python.

b:Use components() to extract the HTML script and div.

c:Pass those to your web framework’s template (Flask or Django).

d:Load BokehJS in the template to make it interactive.
```
